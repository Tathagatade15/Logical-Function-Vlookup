The correct formula is:

The Formula is:

 [=if(B2>=40,"Pass","Fail")].

4. Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

- The most appropriate LOOKUP function to use in Excel for this task is the VLOOKUP function.

 [=VLOOKUP("E103", A1:C5, 3, FALSE)].

Vertical Data Orientation: The data is arranged vertically, meaning the unique identifier (Employee ID) is in the first column, and the information we want to retrieve (Department) is in a subsequent column to the right.

Lookup Process: VLOOKUP searches down the first column of a range (the EmpID column) for a specific value (E103) and returns the value from a corresponding cell in the same row but a specified column (the Department column.
